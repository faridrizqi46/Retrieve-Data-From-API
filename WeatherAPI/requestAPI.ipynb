```python
df = pd.DataFrame()
for i in range(25,30):
    params = {"q":"Tangerang", "dt":f"2021-12-{i}"}
    r = requests.get(url, headers=headers, params=params)
    df = df.append(pd.json_normalize(data=r.json()['forecast']['forecastday'], record_path='hour', 
                            meta=['date','date_epoch']))
```

```python
df = df[['date','date_epoch','time', 'time_epoch', 'temp_c', 'temp_f', 'is_day', 'wind_mph',
       'wind_kph', 'wind_degree', 'wind_dir', 'pressure_mb', 'pressure_in',
       'precip_mm', 'precip_in', 'humidity', 'cloud', 'feelslike_c',
       'feelslike_f', 'windchill_c', 'windchill_f', 'heatindex_c',
       'heatindex_f', 'dewpoint_c', 'dewpoint_f', 'will_it_rain',
       'chance_of_rain', 'will_it_snow', 'chance_of_snow', 'vis_km',
       'vis_miles', 'gust_mph', 'gust_kph', 'condition.text', 'condition.icon',
       'condition.code']]
```
